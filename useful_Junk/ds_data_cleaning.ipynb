# Data Cleaning

#### Importing Necessary Libraries

```python
import pandas as pd
import os
```

##### Merging all the data chunks collected

```python
# df = pd.read_csv('./n_data/n_data_1.csv')

files = [file for file in os.listdir('./n_data')]

all_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./n_data/'+file)
    all_data = pd.concat([all_data, df])

all_data.to_csv("all_data.csv", index = False)   
    
```

```python
all_data = pd.read_csv('all_data.csv')
all_data.count()
```

#### Adding a column

```python
all_data = all_data.drop(['Sr. No'], axis=1)
```
